```python
a=[i for i in range(0,200000)]
df=df.iloc[a].reset_index()
```

```python
import numpy as np
target = np.divide(df[['Production']],df[['Area']])
nans = np.where(np.isnan(target))[0]
zeros = np.where(target==0)[0]
df = df.drop(nans)
target = target.drop(nans)
df = df.drop(zeros)
target = target.drop(zeros)
le = LabelEncoder()
df[['Crop']] = df[['Crop']].apply(le.fit_transform)
le2 = LabelEncoder()
df[['Season']] = df[['Season']].apply(le2.fit_transform)
```

```python
avg = []
for cr in range(0,len(le.classes_)):
    temp = target.loc[df['Crop'] == cr]
    avg.append(np.average(temp))
```
